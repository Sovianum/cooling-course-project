```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import Latex
from IPython.display import Math
pd.set_option('max_colwidth',80)
```

```python
def read_double_compressor_data(file_path):
    return pd.read_csv(file_path, names=[
        'pi', 'pi_factor',
        'G', 'N_e', 'eta',
        'Pi_LPC', 'Pi_HPC',
        'Pi_LPT', 'Pi_HPT',
        'L_HPC', 'L_LPC',
        'L_HPT', 'L_LPT', 'L_FT',
        'Q'])
```

```python
data = read_double_compressor_data('/home/artem/gowork/src/github.com/Sovianum/cooling-course-project/build/data/3n.csv')
```

```python
data
```

```python
tmp = data[data.pi == 20]
plt.plot(tmp.pi_factor, tmp.L_FT / tmp.L_FT.max())
plt.plot(tmp.pi_factor, tmp.Q / tmp.Q.max())
plt.plot(tmp.pi_factor, tmp.eta / tmp.eta.max())
plt.legend(['l_ft', 'q', 'eta'], loc='best')
plt.grid()
plt.ylim([0.965, 1.005])
plt.show()
```

```python
tmp = data[data.pi == 20]
plt.plot(tmp.pi_factor, tmp.Pi_HPT / tmp.Pi_HPT.max())
plt.plot(tmp.pi_factor, tmp.Pi_LPT / tmp.Pi_LPT.max())
plt.plot(tmp.pi_factor, tmp.Pi_HPT * tmp.Pi_LPT / (tmp.Pi_HPT * tmp.Pi_LPT).max())
plt.legend(['high', 'low', 'total'], loc='best')
plt.grid()
plt.show()
```
